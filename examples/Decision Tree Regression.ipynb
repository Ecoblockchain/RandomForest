## Decision Tree Regression (using RandomForest.jl)

Adapted from http://scikit-learn.org/stable/auto_examples/tree/plot_tree_regression.html

A 1D regression for a sine curve with decision tree and random forest.

We can see that that the random forest overfits or learn too much the training data and learn from the noise in it.

```julia
using RandomForest
using PyPlot

# Create a random dataset
srand(42)
X = sort(5 * rand(80))
XX = reshape(X, 80, 1)
y = sin(X)
y[1:5:end] += 3 * (0.5 - rand(16))

# Fit regression model
regr_1 = treeRegressor()
regr_2 = forestRegressor()
fit!(regr_1, XX, y)
fit!(regr_2, XX, y)

# Predict
X_test = 0:0.01:5.0
y_1 = predict(regr_1, hcat(X_test))
y_2 = predict(regr_2, hcat(X_test))

# Plot the results
scatter(X, y, c="k", label="data")
plot(vcat(X_test), y_1, c="g", label="Decision Tree", linewidth=2)
plot(hcat(X_test), y_2, c="b", label="Random Forest", linewidth=2)
xlabel("data")
ylabel("target")
title("Decision Tree Regression")
legend(prop=Dict("size"=>10));
```
